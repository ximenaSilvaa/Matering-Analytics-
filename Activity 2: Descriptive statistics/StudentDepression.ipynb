To improve the visualization and analysis of non-numeric variables, they will be converted into numerical values. For example, in the case of **Dietary Habits**, the categories will be mapped as follows: **Unhealthy** = 0, **Moderate** = 1, **Healthy** = 2, and  **Other** = 3 .


```python

import numpy as np

df = pd.read_csv('student_depression_dataset.csv')

gender_mapping = {'Male': 1, 'Female': 0}
df['Gender'] = df['Gender'].map(gender_mapping).fillna(-1).astype(int)

dietary_mapping = {'Unhealthy': 0, 'Moderate': 1, 'Healthy': 2, 'Other': 3}
df['Dietary Habits'] = df['Dietary Habits'].map(dietary_mapping).fillna(-1).astype(int)

suicide_mapping = {'Yes': 1, 'No': 0}
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map(suicide_mapping).fillna(-1).astype(int)

family_history = {'Yes': 1, 'No': 0}
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].map(suicide_mapping).fillna(-1).astype(int)

df["Financial Stress"] = df["Financial Stress"].replace("?", np.nan) # Now np is defined


df["Financial Stress"] = df["Financial Stress"].astype(float)
```

# Heatmap for Selected Variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('student_depression_dataset.csv')


gender_mapping = {'Male': 1, 'Female': 0}
df['Gender'] = df['Gender'].map(gender_mapping).fillna(-1).astype(int)

dietary_mapping = {'Unhealthy': 0, 'Moderate': 1, 'Healthy': 2, 'Other': 3}
df['Dietary Habits'] = df['Dietary Habits'].map(dietary_mapping).fillna(-1).astype(int)

suicide_mapping = {'Yes': 1, 'No': 0}
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map(suicide_mapping).fillna(-1).astype(int)

family_history = {'Yes': 1, 'No': 0}
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].map(family_history).fillna(-1).astype(int)


df["Financial Stress"] = df["Financial Stress"].replace("?", np.nan).astype(float)

df.to_csv('cleaned_student_depression_dataset.csv', index=False)

df_cleaned = pd.read_csv('cleaned_student_depression_dataset.csv')

selected_columns = [
    'Academic Pressure', 'Study Satisfaction',
    'Dietary Habits',  'Have you ever had suicidal thoughts ?',
    'Work/Study Hours', 'Financial Stress',
    'Family History of Mental Illness', 'Depression'
]


correlation_matrix = df_cleaned[selected_columns].corr()

plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Selected Variables')
plt.show()
```

* Academic Pressure has a moderate positive correlation with Depression (0.47), suggesting that as academic pressure increases, the likelihood of depression also tends to increase.

* Have you ever had suicidal thoughts? shows the strongest positive correlation with Depression (0.55), indicating that students who have had suicidal thoughts are more likely to experience depression.

* Financial Stress also has a noticeable positive correlation with Depression (0.36), highlighting the impact of financial difficulties on mental health.

* Dietary Habits have a weaker but still negative correlation with Depression (-0.21), suggesting that healthier eating habits might be associated with lower depression levels.

* Study Satisfaction has a slight negative correlation with Depression (-0.17), meaning that students who are more satisfied with their studies tend to have lower depression levels.

* Interestingly, Family History of Mental Illness has a very weak correlation with Depression (0.05), which might indicate that other factors have a more direct impact on a student's mental health.

* Work/Study Hours has a small positive correlation with Depression (0.21), possibly reflecting the stress of balancing work and study.
